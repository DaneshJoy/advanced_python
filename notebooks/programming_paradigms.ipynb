# Programming Paradigms

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Procedural Programming

```python
freq = 5
duration = 1
samples = 500
t = np.linspace(0, duration, samples)
signal = np.sin(2 * np.pi * freq * t)
plt.figure(figsize=(4,2))
plt.plot(t, signal, 'b--')
plt.title("Procedural - Sine Wave")
plt.show()
```

## Functional Programming

```python

def create_signal(freq=5, duration=1, samples=500):
    t = np.linspace(0, duration, samples)
    return t, np.sin(2 * np.pi * freq * t)

def plot_signal(t, signal, title="Functional - Signal"):
    plt.figure(figsize=(4,2))
    plt.plot(t, signal, 'g-.')
    plt.title(title)
    plt.show()

t, signal = create_signal()
plot_signal(t, signal)
```

## Object-Oriented Programming (OOP)

```python
class Signal:
    def __init__(self, freq=5, duration=1, samples=500):
        self.freq = freq
        self.duration = duration
        self.samples = samples
        self.t = np.linspace(0, duration, samples)
        self.signal = np.sin(2 * np.pi * freq * self.t)

    def plot(self, title="OOP - Signal"):
        plt.figure(figsize=(4,2))
        plt.plot(self.t, self.signal, 'm:')
        plt.title(title)
        plt.show()

sig = Signal()
sig.plot()
```
