## Functional Programming

```python
def make_signal(signal_type, signal_range):
    x = np.linspace(signal_range[0], signal_range[1], 1000)
    y = None
    if signal_type == 'sin':
        y = np.sin(x)
    elif signal_type == 'cos':
        y = np.cos(x)
    return x, y
```

```python
def visulaize_signal(x, y, title):
    plt.figure(figsize=(4, 2))
    plt.plot(x, y, 'm--')
    plt.title(title)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
```

```python
x1, y1 = make_signal('sin', [0, 2*np.pi])
x2, y2 = make_signal('cos', [0, 2*np.pi])
visulaize_signal(x1, y1, 'Sinus')
visulaize_signal(x2, y2, 'Cosinus')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate time vector
sampling_rate = 250  # Hz
duration = 10  # seconds
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)

# Generate a signal that resembles an EEG signal
# Combine sine waves with different frequencies and amplitudes, and add random noise
eeg_signal = (
    1.5 * np.sin(2 * np.pi * 8 * t) +  # Alpha wave (8-13 Hz)
    0.8 * np.sin(2 * np.pi * 12 * t) + # Alpha wave
    0.5 * np.sin(2 * np.pi * 4 * t) +  # Theta wave (4-7 Hz)
    0.2 * np.sin(2 * np.pi * 1 * t) +  # Delta wave (0.5-3 Hz)
    0.1 * np.random.randn(len(t))      # Random noise
)

# Plot the generated signal
plt.figure(figsize=(10, 4))
plt.plot(t, eeg_signal)
plt.title('Simulated EEG Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
from scipy.signal import butter, filtfilt

def bandpass_filter(signal, lowcut, highcut, sampling_rate, order=5):
    nyquist = 0.5 * sampling_rate
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    filtered_signal = filtfilt(b, a, signal)
    return filtered_signal

filtered_signal = bandpass_filter(eeg_signal, 4, 12, sampling_rate)
```

```python
def visualize_comparison(original_signal, filtered_signal, t, title):
    plt.figure(figsize=(10, 4))
    plt.subplot(2, 1, 1)
    plt.plot(t, original_signal, label='Original Signal', alpha=0.7)
    plt.grid(True)
    plt.subplot(2, 1, 2)
    plt.plot(t, filtered_signal, label='Filtered Signal')
    plt.suptitle(title)
    plt.xlabel('Time (s)')
    plt.grid(True)
    plt.show()

visualize_comparison(eeg_signal, filtered_signal, t, 'Original vs. Filtered Signal')
```
